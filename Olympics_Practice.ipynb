# Final Challenge 

### In every Olympics, which US athlete has won the most total number of medals? Include the athlete descipline.

```python
final=oo[oo.NOC=='USA'].groupby(['Edition','Athlete','Medal']).size().unstack('Medal',fill_value=0)
final['Total'] = final['Gold'] + final['Silver'] + final['Bronze']
final.reset_index(inplace=True)
tu = [group.sort_values('Total',ascending=False)[:1]for year, group in final.groupby('Edition')]
tu
top = pd.DataFrame()
for i in tu:
    top = top.append(i)
top
```
